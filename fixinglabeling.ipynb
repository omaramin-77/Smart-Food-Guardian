```python
import pandas as pd

df = pd.read_csv("FoodFactsCleaned.csv")
LABEL_COL = "nutriscore_letter"
```

```python
print("Original label distribution:")
print(df[LABEL_COL].value_counts().sort_index())
```

```python

# ---- Convert to zero-based ----
df[LABEL_COL] = df[LABEL_COL] + 1

print(df[LABEL_COL].value_counts().sort_index())

# ---- Save cleaned file ----
df.to_csv("FoodFactsCleaned.csv", index=False)

```
